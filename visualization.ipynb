```python
country = [ "india","pakistan","usa","australia"]
growth  = [ 7.7,    3.4,         6.7,   7.2 ]
color   = [ "pink",  "green"   , "red", "yellow" ]
explode = [ 0.2,0.2,0.2,0.2]
# pie chart 
plt.figure(figsize=(10,5))
plt.pie(x=growth,labels=country,colors=color,autopct="%d",
       startangle=90,counterclock=False,explode=explode,
        labeldistance=0.2)
plt.show()
```

```python
import numpy as np
plt.title("NOrmal distribution ")
plt.xlabel("This is x axis")
plt.ylabel("This is y axis")
# plt.legend()
plt.hist(np.random.normal(0,5.5,1200),30,edgecolor='black',color='red')
```

```python
import numpy as np
np.random.seed(43)
plt.title("NOrmal distribution ")
plt.xlabel("This is x axis")
plt.ylabel("This is y axis")
# plt.legend()
plt.hist(np.random.normal(0,5.5,1200),30,edgecolor='black',color='red',orientation='horizontal')
```

```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np 
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(1,1,1,projection='3d')
x = np.random.normal(5.5,1.5,1000)
y = np.random.normal(5.6,3.5,1000)
z = np.random.normal(5.6,4.5,1000)
ax.scatter(x,y,z)
#c=np.linalg.norm([x,y,z],axis=0)
plt.savefig('myplot')
plt.show()
```
